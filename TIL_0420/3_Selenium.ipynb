### Implicit Wait

`.implicitly_wait()`ë¥¼ í™œìš©í•´ì„œ ì•”ì‹œì  ê¸°ë‹¤ë¦¼ì„ ì ìš©í•  ìˆ˜ ìˆë‹¤.

ì£¼ì˜í•  ì ìœ¼ë¡œ, ë°˜ë“œì‹œ í•´ë‹¹ ì‹œê°„ì„ ê¸°ë‹¤ë¦¬ëŠ” ê²ƒì´ ì•„ë‹ˆë¼, ë¡œë”©ì´ ë‹¤ ë ë•Œê¹Œì§€ì˜ í•œê³„ ì‹œê°„ì„ ì˜ë¯¸

````python
# 10ì´ˆ ë™ì•ˆ Implicit Waitì„ ì§„í–‰í•˜ë„ë¡ í•´ì„œ ìŠ¤í¬ë˜í•‘ì´ ì˜ ì´ë£¨ì–´ì§€ë„ë¡ ìˆ˜ì •

from selenium.webdriver.support.ui import WebDriverWait

with webdriver.Chrome(service = Service(ChromeDriverManager().install())) as driver :
    driver.get("https://indistreet.com/live?sortOption=startDate%3ADESC")
    
    # implicitly_wait ë•ë¶„ì—, ì˜¤ë¥˜ê°€ ë‚˜ì§€ ì•ŠìŒ
    driver.implicitly_wait(10)   # ìš”ì²­ì„ ì™„ë²½í•˜ê²Œ ì‹¤í–‰ëœë‹¤ë©´ (ë Œë”ë§ì´ ë‹¤ ë˜ì—ˆë‹¤ë©´,) ë‹¤ìŒìœ¼ë¡œ ë„˜ì–´ê°€ê³ , ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ 10ì´ˆë™ì•ˆ ê¸°ë‹¤ë¦°ë‹¤.
    
    print(driver.find_element(By.XPATH, '//*[@id="__next"]/div/main/div[2]/div/div[4]/div[1]/div[1]/div/a/div[2]/p[1]').text)
````
